```python
import re

# Sample NLP text
text = "Hello! My email is kunal@gmail.com and phone number is 9911380061. Visit https://example.com"

# 1. Tokenization (word extraction)
tokens = re.findall(r'\b\w+\b', text)

# 2. Extract email
emails = re.findall(r'[a-zA-Z0-9._%+-]+@[a-zA-Z.-]+\.[a-zA-Z]{2,}', text)

# 3. Extract phone numbers
phones = re.findall(r'\b\d{10}\b', text)

# 4. Remove URLs (text cleaning)
clean_text = re.sub(r'https?://\S+', '', text)

# 5. Normalize text (lowercase + remove punctuation & digits)
normalized_text = re.sub(r'[^\w\s]', '', clean_text.lower())
normalized_text = re.sub(r'\d+', '', normalized_text)

# Output
print("Tokens:", tokens)
print("Emails:", emails)
print("Phone Numbers:", phones)
print("Clean Text:", normalized_text)
```

```python
import re
from collections import Counter

# Large sample text (generic NLP content)
text = """
Natural Language Processing (NLP) is a field of Artificial Intelligence.
It focuses on enabling machines to understand, analyze, and generate human language.
In 2024, NLP applications include chatbots, search engines, sentiment analysis,
machine translation, speech recognition, and document summarization.
Text data often contains noise such as punctuation, numbers like 12345,
extra spaces, and irrelevant symbols that must be cleaned before modeling.
"""

# 1. Sentence extraction
sentences = re.findall(r'[^.!?]+', text)

# 2. Tokenization (word extraction)
tokens = re.findall(r'\b\w+\b', text.lower())

# 3. Remove numbers
text_no_numbers = re.sub(r'\d+', '', text)

# 4. Remove punctuation
text_no_punct = re.sub(r'[^\w\s]', '', text_no_numbers)

# 5. Normalize text (lowercase + remove extra spaces)
normalized_text = re.sub(r'\s+', ' ', text_no_punct.lower()).strip()

# 6. Keyword search
keyword_found = bool(re.search(r'nlp', normalized_text))

# 7. Remove stop-like short words (length <= 2)
filtered_tokens = [word for word in tokens if len(word) > 2]

# 8. Word frequency
word_frequency = Counter(filtered_tokens)

# Output
print("Sentences:", sentences)
print("\nTotal Tokens:", len(tokens))
print("\nNormalized Text:", normalized_text)
print("\nKeyword 'nlp' Found:", keyword_found)
print("\nTop 5 Most Common Words:", word_frequency.most_common(5))
```

```python
import re
from collections import Counter

# Large sample text
text = """
Natural Language Processing (NLP) is a branch of Artificial Intelligence.
It enables machines to understand and generate human language.
In modern systems, text data contains punctuation, numbers like 12345,
extra spaces, and multiple sentences that need preprocessing.
In 2024, NLP applications include chatbots, search engines, sentiment analysis,
machine translation, speech recognition, and document summarization.
"""

# -------------------- REGEX PATTERNS --------------------
sentence_pattern = r'[^.!?]+'
token_pattern = r'\b\w+\b'
number_pattern = r'\d+'
punctuation_pattern = r'[^\w\s]'
space_pattern = r'\s+'
keyword_pattern = r'\bnlp\b'

# -------------------- PRINT REGEX PATTERNS --------------------
print("Sentence Regex:", sentence_pattern)
print("Token Regex:", token_pattern)
print("Number Regex:", number_pattern)
print("Punctuation Regex:", punctuation_pattern)
print("Whitespace Regex:", space_pattern)
print("Keyword Regex:", keyword_pattern)

# -------------------- NLP OPERATIONS --------------------

# 1. Sentence extraction
sentences = re.findall(sentence_pattern, text)

# 2. Tokenization
tokens = re.findall(token_pattern, text.lower())

# 3. Remove numbers
text_no_numbers = re.sub(number_pattern, '', text)

# 4. Remove punctuation
text_no_punct = re.sub(punctuation_pattern, '', text_no_numbers)

# 5. Normalize text (lowercase + remove extra spaces)
normalized_text = re.sub(space_pattern, ' ', text_no_punct.lower()).strip()

# 6. Keyword search
keyword_found = bool(re.search(keyword_pattern, normalized_text))

# 7. Remove short words (length â‰¤ 2)
filtered_tokens = [word for word in tokens if len(word) > 2]

# 8. Word frequency
word_frequency = Counter(filtered_tokens)

# -------------------- OUTPUT --------------------
print("\nNumber of Sentences:", len(sentences))
print("Total Tokens:", len(tokens))
print("Normalized Text:", normalized_text)
print("Keyword 'nlp' Found:", keyword_found)
print("Top 5 Most Common Words:", word_frequency.most_common(5))
```

```python
from nltk import Tree

# Parse Tree 1: PP attached to the VERB (instrument interpretation)
tree1 = Tree('S', [
    Tree('NP', ['I']),
    Tree('VP', [
        'saw',
        Tree('NP', ['the', 'man']),
        Tree('PP', [
            'with',
            Tree('NP', ['the', 'telescope'])
        ])
    ])
])

# Parse Tree 2: PP attached to the NOUN (modifier interpretation)
tree2 = Tree('S', [
    Tree('NP', ['I']),
    Tree('VP', [
        'saw',
        Tree('NP', [
            'the',
            'man',
            Tree('PP', [
                'with',
                Tree('NP', ['the', 'telescope'])
            ])
        ])
    ])
])

print("Parse Tree 1: PP attached to Verb")
tree1.pretty_print()

print("\nParse Tree 2: PP attached to Noun")
tree2.pretty_print()
```

```python
from nltk import Tree

# Parse Tree 1: Prepositional Phrase attached to the Verb Phrase
tree1 = Tree('Sentence', [
    Tree('Noun_Phrase', ['I']),
    Tree('Verb_Phrase', [
        'saw',
        Tree('Noun_Phrase', ['the', 'man']),
        Tree('Prepositional_Phrase', [
            'with',
            Tree('Noun_Phrase', ['the', 'telescope'])
        ])
    ])
])

# Parse Tree 2: Prepositional Phrase attached to the Noun Phrase
tree2 = Tree('Sentence', [
    Tree('Noun_Phrase', ['I']),
    Tree('Verb_Phrase', [
        'saw',
        Tree('Noun_Phrase', [
            'the',
            'man',
            Tree('Prepositional_Phrase', [
                'with',
                Tree('Noun_Phrase', ['the', 'telescope'])
            ])
        ])
    ])
])

print("Parse Tree 1: Prepositional Phrase attached to Verb Phrase")
tree1.pretty_print()

print("\nParse Tree 2: Prepositional Phrase attached to Noun Phrase")
tree2.pretty_print()
```

```python
import nltk
from nltk import CFG
from nltk.parse import RecursiveDescentParser, ChartParser

# Define a simple grammar
grammar = CFG.fromstring("""
Sentence -> Noun_Phrase Verb_Phrase
Noun_Phrase -> 'I' | 'the' 'man'
Verb_Phrase -> 'saw' Noun_Phrase
""")

sentence = "I saw the man".split()

# ---------- Top-Down Parsing ----------
print("Top-Down Parsing:")
top_down = RecursiveDescentParser(grammar)

for tree in top_down.parse(sentence):
    tree.pretty_print()

# ---------- Bottom-Up Parsing ----------
print("\nBottom-Up Parsing:")
bottom_up = ChartParser(grammar)

for tree in bottom_up.parse(sentence):
    tree.pretty_print()
```
